```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic ploting

```python
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]

#resize the graph
plt.figure(figsize = (5, 3), dpi = 100)
plt.plot(x, y, marker="x")
plt.title("Linear Graph", fontdict = {"fontname": "Times New Roman", "fontsize": 20})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#making some x-ticks or y-ticks
plt.xticks([1, 3, 5, 7])
plt.yticks([2, 6, 10])
x2 = np.arange(0, 4.5, 0.5)

plt.plot(x2, x2**2, 'r')
plt.show()
plt.savefig("mygraph.png", dpi = 300)
```

## Bar Charts

```python
labels  = ['A', 'B', 'C']
values = [2 ,4, 8]

bars = plt.bar(labels, values)

plt.figure(figsize = (8, 4))
bars[2].set_hatch("*")
plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")
gas
```

```python
plt.plot(gas.Year, gas.USA, '.-', label = "USA")
plt.plot(gas.Year, gas.Canada, '.-', label = "Canada")
plt.plot(gas.Year, gas["South Korea"], '.-', label = "South Korea")
plt.xlabel("Year")
plt.ylabel("Us Dollars")
plt.xticks(gas.Year[::3])
plt.title("Gas prices over time in USD", fontdict = {"fontname": "Times New Roman", "fontsize": 15,"fontweight": 'bold'})
plt.legend()
plt.show()
```

## Ploting all countries

```python
plt.figure(figsize = (10, 7))
for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], '.-', label = country)
plt.xlabel("Year")
plt.ylabel("Us Dollars")
plt.xticks(gas.Year[::3])
plt.title("Gas prices over time in USD")
plt.legend()
plt.savefig("gaz_evollution.png", dpi = 300)

plt.show()  
```
