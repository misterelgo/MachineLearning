# Pandas file handling

```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
df
```

```python
print(df.head(10))
```

```python
print(df.columns)
```

```python
#read a specific location with iloc
print(df.iloc[0])
```

```python
#read a specific value based on a filter the loc keyword
print(df.loc[df["Type 2"] == "Dragon"])
```

## Sorting and describing data

```python
df.describe()
```

```python
df.sort_values(["Name", "Type 1"], ascending = [False, True])
```

## Making chnages to the data

```python
df["Totals"] = df["Attack"] + df["Defense"]
df.head(5)
```

```python
#Deleting a column
df = df.drop(columns = ["Totals"])
df.head()
```

```python
df["Totals"] = df.iloc[:, 4:10].sum(axis = 1)
df.head()
```

## Saving the file after modification

```python
#df.to_csv("modified.csv", index = False)
#df.to_excel("modified.xlsx", index = False)
df.to_csv("modified.txt", index = False, sep = " ")
```

## Data filtering

```python
#df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["Speed"] < 40)].to_csv("filterd_data.csv", index = False)
#df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison

#We can also use regular expresions 
#flags = re.I is for ignoring case sensitivity
import re
df.loc[df["Name"].str.contains('der|Aur', flags = re.I, regex = True)]
```

## Filter with conditions

```python
#~ means 'not'
df.loc[~df["Name"].str.contains("aur")]
```

## Changing data on condition

```python
#df.loc[df["Type 1"] == "Grass", "Type 1"] = "Not Grass"
#We can change multiple colum at the same time
df.loc[df["Type 1"] == "Not Grass", ["Generation", "Legendary"]] = [0, True]
df
```

## Aggregate Statistics with (GroupBy) function

```python
#df.groupby(["Type 1"]).mean().sort_values("Defense", ascending = False)
df.groupby(["Type 1"]).count().sort_values("Defense", ascending = False)
```
