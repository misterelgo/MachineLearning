## Discovering patterns with Algorithm of sequence mining

```python
#The global goal of the algorithm of sequence mining is to
#discover every pattern contained in the database, and evaluate their support
```

```python
#Discovering patterns of elementary length
#We consider a pattern is elementary when it got at least 2 captors on simultaneously
patternList = []
for seq in range(len(sequence.index)):
    listEvent = []
    for event in sequence.iloc[seq]:
        listEvent.append(event)
    eventCand = []
    i = 0
    while (i <= len(listEvent)-1):
        if listEvent[i] == 1:
            eventCand.append(listEvent.index(listEvent[i])+1)
            listEvent[i] = 0
        i+=1
    if(len(eventCand) >= 2):
        patternList.append(eventCand)     
    
print(patternList)
```
